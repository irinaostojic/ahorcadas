```python
import random  # libreria para combinaciones aleatorias

# Palabras de terror escogidas
palabras_terror = [
    "espectro", "cripta", "ataud", "tinieblas", "cementerio", "maldicion",
    "fantasma", "exorcismo", "calavera", "posesion", "hechizo", "bruja",
    "ultratumba", "demonio", "sepulcro"
]

# Diccionario de frases de terror
frases_terror = {
    "bienvenida": [
        "Bienvenida... al ultimo juego de tu vida.",
        "La noche es oscura. Bienvenida, esta puede ser la ultima palabra que adivines.",
        "Has entrado. ¡Bienvenida! No podras salir hasta que ganes."
    ],
    "acierto": [
        "Si... pero ellos ya saben que estas aqui.",
        "Acertaste... pero algo se mueve en las sombras.",
        "Muy bien... por ahora sobrevives."
    ],
    "fallo": [
        "Has fallado... y la soga aprieta un poco mas.",
        "No... y algo se arrastra hacia ti.",
        "Ese error te acerca al final."
    ],
    "victoria": [
        "Has sobrevivido... pero no por mucho tiempo.",
        "Ganaste... o eso crees.",
        "La palabra era tu salvacion... esta vez."
    ],
    "derrota": [
        "Has perdido... y ahora eres uno de ellos.",
        "No lograste escapar... la oscuridad te envuelve.",
        "Caiste... y nadie vendra a ayudarte."
    ]
}

# Seleccion aleatoria de palabra
palabra = random.choice(palabras_terror)

# Mensajes iniciales
print(random.choice(frases_terror["bienvenida"]))
print(f"Tu palabra contiene {len(palabra)} letras: {'_ ' * len(palabra)}")
print(palabra) #la imprimimos para ver si nos funciona el código

# Etapas del juego del ahorcado
ahorcado = [
    """
     ------
     |    |
     |
     |
     |
    _|_
    """,
    """
     ------
     |    |
     |    O
     |
     |
    _|_
    """,
    """
     ------
     |    |
     |    O
     |    |
     |
    _|_
    """,
    """
     ------
     |    |
     |    O
     |   /|
     |
    _|_
    """,
    """
     ------
     |    |
     |    O
     |   /|\\
     |
    _|_
    """,
    """
     ------
     |    |
     |    O
     |   /|\\
     |   /
    _|_
    """,
    """
     ------
     |    |
     |    O
     |   /|\\
     |   / \\
    _|_
    """
]

# Variables de control del juego
letras_correctas = set()
letras_incorrectas = set()
letras_usadas = set()
intentos = len(ahorcado) - 1
fallos = 0

# Configuracion de jugadores y marcador de puntos para cada jugador
jugadores = ["JUGADOR 1", "JUGADOR 2"]
puntos = {jugadores[0]: 0, jugadores[1]: 0}  # marcador inicial
turno = 0  # empieza el Jugador 1
print(f"Tienes {intentos} intentos para acertar la palabra. Vivir o morir. Tu decides.")

# Bucle principal del Juego del Ahorcado
while True:

    # Nos muestra el dibujo inicial del ahorcado y nos muestra el turno del jugador
    jugador_actual = jugadores[turno]
    print(f"\nTURNO DE {jugador_actual}")
    print(ahorcado[fallos])

    # Mostrar estado actual de la palabra
    for letra in palabra:
        if letra.lower() in letras_correctas:
            print(letra, end=" ")
        else:
            print("_", end=" ")
    print()

    print(f"Letras usadas: {', '.join(sorted(letras_usadas)) if letras_usadas else 'Ninguna'}")  # para saber las letras que vamos usando

    # Introduccion de letra
    letra = input("Introduce una letra: ").strip().lower()

    # Validaciones a las entradas de letra
    if len(letra) != 1 or not letra.isalpha():
        print(f"{letra} no es un caracter valido. Prueba de nuevo")
        continue

    if letra in letras_usadas:
        print(f"Ya has usado la letra '{letra}'. Prueba con otra.")
        continue

    letras_usadas.add(letra)

    # Comprobacion de letra
    if letra in palabra.lower():
        letras_correctas.add(letra)
        print(f"{random.choice(frases_terror['acierto'])}. La '{letra}' esta en la palabra buscada")
    else:
        letras_incorrectas.add(letra)
        fallos += 1
        print(f"{random.choice(frases_terror['fallo'])}. La '{letra}' no esta en la palabra buscada")
        print(f"Te quedan {intentos - fallos} intentos")
        turno = 1 - turno  # cambiar turno solo si falla
        if fallos < intentos:
            print(f"Ahora es el turno de {jugadores[turno]}")

    # Condicion de aciertos en el Juego
    if letras_correctas >= set(palabra.lower()):
        print(random.choice(frases_terror["victoria"]))
        print(f"¡{jugadores[turno]} ERES EL GANADOR DEL JUEGO DEL AHORCADO! La palabra era: {palabra}")
        puntos[jugadores[turno]] += 1  # va sumando puntos al que acierta
        print(f"Marcador actual: {jugadores[0]}: {puntos[jugadores[0]]} | {jugadores[1]}: {puntos[jugadores[1]]}")
        break

    # Condicion de fallos en el Juego
    if fallos == intentos:
        print(f"{random.choice(frases_terror['derrota'])}. La palabra era: {palabra}")
        ganador = jugadores[1 - turno]
        print(f"¡{ganador} ES EL GANADOR DEL JUEGO DEL AHORCADO!")
        puntos[ganador] += 1  # sumar punto al ganador
        print(f"El Marcador Final es: {jugadores[0]}: {puntos[jugadores[0]]} | {jugadores[1]}: {puntos[jugadores[1]]}")
        break
```
