```python
import random

# Lista de palabras
palabras = [
    'l√≥gica', 'proactividad', 'creatividad',
    'resiliencia', 'comunicaci√≥n', 'python',
    'programaci√≥n', 'datos'
]

# Elegir palabra al azar
palabra = random.choice(palabras)

print("¬°¬°¬°BIENVENIDOS AL JUEGO DEL AHORCADO PARA 2 JUGADORES!!!")
# Solo para pruebas, puedes quitarlo despu√©s
print(f"(Pista: la palabra es '{palabra}')")
print(f"La palabra contiene {len(palabra)} letras: {'_ ' * len(palabra)}")

# Etapas del ahorcado
ahorcado = [
    """
     ------
     |    |
     |
     |
     |
    _|_
    """,
    """
     ------
     |    |
     |    O
     |
     |
    _|_
    """,
    """
     ------
     |    |
     |    O
     |    |
     |
    _|_
    """,
    """
     ------
     |    |
     |    O
     |   /|
     |
    _|_
    """,
    """
     ------
     |    |
     |    O
     |   /|\\
     |
    _|_
    """,
    """
     ------
     |    |
     |    O
     |   /|\\
     |   /
    _|_
    """,
    """
     ------
     |    |
     |    O
     |   /|\\
     |   / \\
    _|_
    """
]

# Variables del juego
letras_correctas = set()
letras_incorrectas = set()
letras_usadas = set()
intentos = len(ahorcado) - 1
fallos = 0

# Configuraci√≥n de jugadores
jugadores = ["Jugador 1", "Jugador 2"]
turno = 0  # empieza el Jugador 1

print(f"Tienen {intentos} intentos entre los dos para adivinar la palabra.\n")

# Bucle principal
while True:
    print(ahorcado[fallos])

    # Mostrar estado actual de la palabra
    for letra in palabra:
        if letra.lower() in letras_correctas:
            print(letra, end=" ")
        else:
            print("_", end=" ")
    

    # Mostrar informaci√≥n del turno
    jugador_actual = jugadores[turno] #Nuevo
    print(f"Turno de {jugador_actual}") #Nuevo
    print(f"Letras usadas: {', '.join(sorted(letras_usadas)) if letras_usadas else 'Ninguna'}") #Nuevo

    # Pedir letra
    letra = input("Introduce una letra: ").strip().lower()

    # Validar entrada
    if len(letra) != 1 or not letra.isalpha():
        print("‚ö†Ô∏è Entrada inv√°lida. Introduce una sola letra.")
        continue

    if letra in letras_usadas:
        print("‚ö†Ô∏è Esa letra ya fue usada. Elige otra.")
        continue

    letras_usadas.add(letra)

    # Comprobar si est√° en la palabra
    if letra in palabra:
        letras_correctas.add(letra)
        print(f"‚úÖ ¬°Bien {jugador_actual}! La letra '{letra}' est√° en la palabra.")
    else:
        letras_incorrectas.add(letra)
        fallos += 1
        print(f"La letra '{letra}' no est√° en la palabra")
        print(f"Te quedan {(len(ahorcado)-1 - fallos)} intentos")
        print(f"‚ùåLetras incorrectas: {', '.join(sorted(letras_incorrectas))}")
        

    # Comprobar victoria
    if letras_correctas == set(palabra):
        print(f"\nüéâ ¬°FELICIDADES {jugador_actual}! Has adivinado la palabra: {palabra}")
        break

    # Comprobar derrota
    if fallos == intentos:
        print(f"\nüíÄ ¬°GAME OVER! La palabra era: {palabra}")
        print(ahorcado[fallos])
        break

    # Cambiar turno
    turno = 1 - turno
```
