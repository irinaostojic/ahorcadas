```python
import random # librería para combinaciones aleatorias

# Palabra de terror escogidas para después escoger una palabra aleatoria con python
palabras_terror = [
    "espectro", "cripta", "ataúd", "tinieblas", "cementerio", "maldición",
    "fantasma", "exorcismo", "calavera", "posesión", "hechizo", "bruja",
    "ultratumba", "demonio", "sepulcro"]     
       
# Diccionario de frases de terror para el print de bienvenida, acierto, fallo, victoria y derrota
frases_terror = {
    "bienvenida": [
        "Bienvenida... al último juego de tu vida.",
        "La noche es oscura. Bienvenida, esta puede ser la última palabra que adivines.",
        "Has entrado. ¡Bienvenida! No podrás salir hasta que ganes."
    ],
    "acierto": [
        "Sí... pero ellos ya saben que estás aquí.",
        "Acertaste... pero algo se mueve en las sombras.",
        "Muy bien... por ahora sobrevives."
    ],
    "fallo": [
        "Has fallado... y la soga aprieta un poco más.",
        "No... y algo se arrastra hacia ti.",
        "Ese error te acerca al final."
    ],
    "victoria": [
        "Has sobrevivido... pero no por mucho tiempo.",
        "Ganaste... o eso crees.",
        "La palabra era tu salvación... esta vez."
    ],
    "derrota": [
        "Has perdido... y ahora eres uno de ellos.",
        "No lograste escapar... la oscuridad te envuelve.",
        "Caíste... y nadie vendrá a ayudarte."
    ]
}

# Diccionario de vocales acentuadas y sin acentuar
acentos = {
    'á': 'a', 'é': 'e', 'í': 'i', 'ó': 'o', 'ú': 'u',
    'à': 'a', 'è': 'e', 'ì': 'i', 'ò': 'o', 'ù': 'u'
}

# Selección aletoaria de palabra
palabra = random.choice(palabras_terror)

# Mensajes iniciales
print(random.choice(frases_terror["bienvenida"]))
print(palabra)  # Solo para pruebas
print(f"Tu palabra contiene {len(palabra)} letras: {'_ ' * len(palabra)}")

# Etapas del juego del ahorcado
ahorcado = [
    """
     ------
     |    |
     |
     |
     |
    _|_
    """,
    """
     ------
     |    |
     |    O
     |
     |
    _|_
    """,
    """
     ------
     |    |
     |    O
     |    |
     |
    _|_
    """,
    """
     ------
     |    |
     |    O
     |   /|
     |
    _|_
    """,
    """
     ------
     |    |
     |    O
     |   /|\\
     |
    _|_
    """,
    """
     ------
     |    |
     |    O
     |   /|\\
     |   /
    _|_
    """,
    """
     ------
     |    |
     |    O
     |   /|\\
     |   / \\
    _|_
    """
]

# Variables de control del juego
letras_correctas = set()
letras_incorrectas = set()
letras_usadas = set()
intentos = len(ahorcado) - 1
fallos = 0
print(f"Tienes {intentos} intentos para acertar la palabra. Vivir o morir. Tú decides.")
```

```python

while True:
    print(ahorcado[fallos])

    # Bucle inicial de iteración. Aquí nos muestra la letra o guión. También aquí le decimos a python que nos reconozca la vocal aunque nos equivoquemos en el acento
    for letra_original in palabra:
        letra_sin_acento = acentos.get(letra_original.lower(), letra_original.lower())
        if letra_sin_acento in letras_correctas:
            print(letra_original, end=" ")
        else:
            print("_", end=" ")
    print()

    letra = input("Introduce una letra: ").strip().lower()
    letra = acentos.get(letra, letra)  # normalizamos la entrada

    if len(letra) != 1 or not letra.isalpha():
        print(f"{letra} no es un carácter válido. Prueba de nuevo")
        continue

    if letra in letras_usadas:
        print(f"Ya has usado la letra '{letra}'. Prueba con otra.")
        continue

    letras_usadas.add(letra)

    # Comprobamos letra quitando acentos a la palabra
    if letra in [acentos.get(c, c) for c in palabra.lower()]:
        letras_correctas.add(letra)
        print(random.choice(frases_terror["acierto"]))
        print(f"La {letra} está en la palabra buscada")
    else:
        letras_incorrectas.add(letra)
        fallos += 1
        print(random.choice(frases_terror["fallo"]))
        print(f"La¨{letra} no está en la palabra buscada")
        print(f"Te quedan {(len(ahorcado) - 1 - fallos)} intentos")


    # Ganamos en el juego
    if letras_correctas >= set(acentos.get(c, c) for c in palabra.lower()):
        print(random.choice(frases_terror["victoria"]))
        print(f"Has acertado la palabra : {palabra}")
        break

    # Perdemos en el juego. Hemos consumido todas las vidas
    if fallos == intentos:
        print(random.choice(frases_terror["derrota"]))
        print(f"La palabra era: {palabra}")
        print(ahorcado[fallos])
        break
```
