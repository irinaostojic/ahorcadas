```python
import random # necesitamos la librería para combinaciones aleatorias

palabras = ['lógica', 'proactividad', 'creatividad', 'resiliencia', 'comunicación', 'python', 'programación', 'datos'] #hay que tener cuidado más adelante de las palabras que tienen acento
      
palabra = random.choice(palabras) #python me saca una aleatoria

print("¡¡¡BIENVENIDA AL JUEGO DEL AHORCADO!!!")    #mensaje de entrada para empezar el juego
print(palabra) #quiero ver la palabra qué es solo para ir viendo qué pasa y probando 
print(f"Tu palabra contiene {len(palabra)} letras: {'_ ' * len(palabra)}")  #nos dice las letras que tiene la palabra a adivinar


# Variables que necesitamos para empezar nuestro juego

ahorcado = [
    """
     ------
     |    |
     |
     |
     |
    _|_
    """,
    """
     ------
     |    |
     |    O
     |
     |
    _|_
    """,
    """
     ------
     |    |
     |    O
     |    |
     |
    _|_
    """,
    """
     ------
     |    |
     |    O
     |   /|
     |
    _|_
    """,
    """
     ------
     |    |
     |    O
     |   /|\\
     |
    _|_
    """,
    """
     ------
     |    |
     |    O
     |   /|\\
     |   /
    _|_
    """,
    """
     ------
     |    |
     |    O
     |   /|\\
     |   / \\
    _|_
    """] #las etapas del juego del ahorcado

# necesitamos listas vacías tanto de las letras correctas como de las incorrectas para ir llenándolas
# es más sencillo utilizar set, así vemos si un conjunto de letras es subconjunto de otro
letras_correctas = set()
letras_incorrectas = set()
letras_usadas = set() # podemos poner otro conjunto adicional para saber que hemos puesto ya
intentos = len(ahorcado)-1  #los intentos los definimos según los elementos de la lista del ahorcado memos 1, el primero es la estructura donde cuelga
fallos = 0 #empezamos en cero y el máximo serán los intentos definidos
print(f"Tienes {intentos} intentos para acertar la palabra. ¡Ánimo!")
```

```python
while True:
    
    print(ahorcado[fallos])
    
    #1. Primer bucle:
    # Si la letra está en letras correctas (letras correctas es una lista que va a ir rellenándose en cada nueva entrada al bucle, hay que crearla vacía en un principio para poder ir llenándola)
    # hay que entender el for del inicio en su conjunto, es decir aunque se parta de una lista vacía de palabras correctas, está todo el rato llenándose en cada entrada nueva al bucle
   
    for letra in palabra:
        if letra.lower() in letras_correctas:
            print(letra, end=" ")

        else:
            print("_", end=" ")


    # Pedimos al usuario que introduzca la letra y hacemos validaciones
    letra = input("Introduce una letra: ").strip().lower()

    if len(letra) != 1 or not letra.isalpha(): 
        print(f"{letra} no es un carácter válido. Prueba de nuevo")
        continue
        

    if letra in letras_usadas:
        print(f"Ya has usado esa {letra} Prueba con otra")
        continue
    
    letras_usadas.add(letra) # agregamos la letra a conjunto de usadas


    #2. Comprobar si la letra está en la palabra
    if letra in palabra:
        letras_correctas.add(letra)    #si la letra que hemos dicho está en la palabra, se añade a la lista de letras correctas
        print(f"¡Bien! La letra '{letra}' está en la palabra.")
        
    else:
        letras_incorrectas.add(letra)  # si no está en la lista de correctas, va a incorrectas y se suma un fallo
        fallos += 1
        print(f"La letra '{letra}' no está en la palabra")
        print(f"Te quedan {(len(ahorcado)-1 - fallos)} intentos")
        print(f"Letras incorrectas: {', '.join(sorted(letras_incorrectas))}") #las vamos viendo impresas
        

# Ahora queremos saber cómo va evolucionando el juego, si pierde o si gana

    if letras_correctas == set(palabra):
        print(f"¡ENHORABUENA! HAS ACERTADO LA PALABRA: {palabra}")
        break  #sales del bucle cuando aciertas la palabra

    # Comprobar si ha perdido
    if fallos == intentos:
        print(f"¡GAME OVER! La palabra era: {palabra}")
        print(ahorcado[fallos])
        break #sales del bucle cuando consumes todas tus vidas
```
