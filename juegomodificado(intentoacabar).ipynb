```python
import random        #librer√≠a que necesito para hacer las combinaciones aleatorias

palabras = ['l√≥gica', 'proactividad', 'creatividad', 'resiliencia', 'comunicaci√≥n']
palabra = random.choice(palabras)      # le digo a python que me saque una palabra aleatoria de mi lista
print(palabra)         #quiero ver la palabra qu√© es solo para ir viendo qu√© pasa y probando 

print("=== Juego del Ahorcado ===")    #mensaje de entrada para empezar el juego

ahorcado = [
    """
     +---+
     |   |
         |
         |
         |
         |
    =========
    """,
    """
     +---+
     |   |
     O   |
         |
         |
         |
    =========
    """,
    """
     +---+
     |   |
     O   |
     |   |
         |
         |
    =========
    """,
    """
     +---+
     |   |
     O   |
    /|   |
         |
         |
    =========
    """,
    """
     +---+
     |   |
     O   |
    /|\\  |
         |
         |
    =========
    """,
    """
     +---+
     |   |
     O   |
    /|\\  |
    /    |
         |
    =========
    """,
    """
     +---+
     |   |
     O   |
    /|\\  |
    / \\  |
         |
    =========
    """
]     # es una simple lista de dibujos con el ahorcado. Si queremos caritas tenemos que hacer tupla

intentos = len(ahorcado) - 1    # 7 dibujitos, 1 por defecto, se lo resto
contador_fallos = 0      # le ponemos un contador de intentos fallidos, como mucho tendremos 6

letras_usadas = set() # conjunto de las letras que vamos usando
letras_correctas = set() # conjunto de las letras que vamos acertando

# creo que es m√°s correcto utilizar conjuntos, pues no necesitamos ning√∫n orden
print(f"Tienes {intentos} intentos fallidos para adivinar la palabra.")
while contador_fallos < intentos and set(palabra) != letras_correctas:
    # mostrar el dibujo correspondiente a los fallos que llevamos
    print(ahorcado[contador_fallos])
    
    # mostrar progreso de la palabra (letra acertada o guion bajo)
    progreso = ""
    for letra in palabra:
        if letra in letras_correctas:
            progreso += letra + " "
        else:
            progreso += "_ "
    print("Palabra:", progreso.strip())
    
    # mostrar letras ya usadas
    print("Letras usadas:", " ".join(sorted(letras_usadas)))
    
    # pedir letra al jugador
    intento = input("Introduce una letra: ").lower()
    
    # validar entrada
    if len(intento) != 1 or not intento.isalpha():
        print("‚ö†Ô∏è Por favor, introduce solo UNA letra.\n")
        continue
    
    if intento in letras_usadas:
        print("Ya has usado esa letra, prueba otra.\n")
        continue
    
    # a√±adir la letra al conjunto de usadas
    letras_usadas.add(intento)
    
    # comprobar si la letra est√° en la palabra
    if intento in palabra:
        letras_correctas.add(intento)
        print("‚úÖ ¬°Letra correcta!\n")
    else:
        contador_fallos += 1
        print("‚ùå Letra incorrecta.\n")

# salida del bucle: comprobar si gan√≥ o perdi√≥
if set(palabra) == letras_correctas:
    print("üéâ ¬°Felicidades! Has adivinado la palabra:", palabra)
else:
    print(ahorcado[contador_fallos])
    print(f"üíî Has perdido. La palabra era: {palabra}")
```
