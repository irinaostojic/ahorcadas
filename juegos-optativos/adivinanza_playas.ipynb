#### Juego de Adivinanza de Playas españolas:

Dos jugadores se van turnando para adivinar la ciudad en la que se encuentra la playa. 

Se van turnando indistintamente de si fallan o pierden, el que primero llegue, por ejemplo, a 3 puntos (acierte 3 playas), gana el juego. No puede haber empates, el que primero llegue a 3, gana.

Los jugadores pueden pedir pistas.

Con "salir" se sale del bucle del juego.

(se ha resuelto con una clase, métodos, e instancias de clase)



```python
import random
import unicodedata

class AdivinarPlayas:
    def __init__(self):
        self.playas = [
            "Playa de Ses Illetes", "Playa de Bolonia", "Playa de San Juan", 
            "Playa de Maspalomas", "Playa de la Barceloneta", "Playa de Alcudia", "La Concha",
            "Playa de las Canteras", "Playa de la Victoria", "Playa de Amadores",
            "Playa de la Lanzada", "Playa de Mijares", "Playa de Zahara de los Atunes",
            "Playa de Carnota", "Playa de La Herradura", "Playa de las Catedrales", "Playa de la Malvarrosa",
            "Playa de Covachos", 
        ]

        self.ciudades = {
            "La Concha": "San Sebastián",
            "Playa de las Catedrales": "Ribadeo",
            "Playa de Ses Illetes": "Formentera",
            "Playa de Bolonia": "Tarifa",
            "Playa de la Malvarrosa": "Valencia",
            "Playa de Zahara de los Atunes": "Zahara",
            "Playa de la Victoria": "Cádiz",
            "Playa de la Barceloneta": "Barcelona",
            "Playa de Maspalomas": "Maspalomas",
            "Playa de Alcudia": "Mallorca",
            "Playa de la Lanzada": "Sanxenxo",
            "Playa de San Juan": "Alicante",
            "Playa de Amadores": "Gran Canaria",
            "Playa de Mijares": "Castellón",
            "Playa de Carnota": "Carnota",
            "Playa de las Canteras": "Las Palmas",
            "Playa de La Herradura": "Almuñécar",
            "Playa de Covachos": "Santander",
        }

        self.pistas = {
            "La Concha": "Playa icónica del País Vasco, formando una bahía.",
            "Playa de las Catedrales": "Sus enormes arcos rocosos parecen catedrales naturales.",
            "Playa de Ses Illetes": "Arena clara y aguas turquesa, en una pequeña isla balear.",
            "Playa de Bolonia": "Frente a un antiguo yacimiento romano, arena fina y dunas.",
            "Playa de la Malvarrosa": "Playa urbana muy popular en la capital del Turia.",
            "Playa de Zahara de los Atunes": "Conocida por su pesca de atún y su costa tranquila.",
            "Playa de la Victoria": "Amplia playa urbana de arena dorada al sur de España.",
            "Playa de la Barceloneta": "Muy concurrida, perfecta para pasear por el paseo marítimo.",
            "Playa de Maspalomas": "Famosa por sus dunas y su ambiente turístico en las Islas Canarias.",
            "Playa de Alcudia": "Playa larga y de aguas poco profundas en el norte de una isla balear.",
            "Playa de la Lanzada": "Ideal para practicar surf en la costa gallega.",
            "Playa de San Juan": "Playa urbana de gran tamaño con paseos y chiringuitos.",
            "Playa de Amadores": "Arena dorada y agua tranquila, perfecta para nadar.",
            "Playa de Mijares": "Arena suave y mar en calma en la provincia de Castellón.",
            "Playa de Carnota": "Una de las playas más largas de Galicia, salvaje y extensa.",
            "Playa de las Canteras": "Ubicada en las islas, con buen oleaje para surf.",
            "Playa de La Herradura": "Playa de Andalucía, con aguas calmadas y acantilados.",
            "Playa de Covachos": "Pequeña playa rodeada de acantilados en Cantabria.",
        }

        self.puntos = {"Jugador 1": 0, "Jugador 2": 0}
        self.jugador_actual = "Jugador 1"

    def alternar_turno(self): # Método para que se turnen los jugadores
        print(f"\n--- Fin del turno de {self.jugador_actual} ---") #termina el turno del jugador actual

        #Cambia al otro jugador
        if self.jugador_actual == "Jugador 1":  #si está jugando el 1, cambia al 2
            self.jugador_actual = "Jugador 2"
        else:
            self.jugador_actual = "Jugador 1"

        print(f"--- Comienza el turno de {self.jugador_actual} ---\n") #Empieza el turno del nuevo jugador


    def mostrar_puntos(self): #Método para mostrar las puntuaciones
        print("\nPuntaje actual:")
        for jugador, punto in self.puntos.items(): #recorre el diccionario de puntos para cada jugador
            print(f"{jugador}: {punto}")


    def jugar_ronda(self, playa): #Juego
        while True:
            respuesta = input(f"{self.jugador_actual}, ¿En qué ciudad se ubica la playa '{playa}'? ").strip()
            respuesta_normalizada = unicodedata.normalize('NFKD', respuesta).encode('ASCII', 'ignore').decode('utf-8').lower()
            ciudad_correcta = unicodedata.normalize('NFKD', self.ciudades[playa]).encode('ASCII', 'ignore').decode('utf-8').lower()

            if respuesta_normalizada == "salir":
                return False   #salir del juego
            
            elif respuesta_normalizada == "pista": #dame pista
                print(f"Pista: {self.pistas[playa]}")
                continue

            elif respuesta_normalizada == ciudad_correcta:
                print("¡RESPUESTA CORRECTA!")
                self.puntos[self.jugador_actual] += 1
            else:
                print("RESPUESTA INCORRECTA.")

            self.mostrar_puntos()

            # Termina el juego si alguien llega a 7 puntos
            if self.puntos[self.jugador_actual] >= 3:   #Por ejemplo 4 puntos
                print(f"\n¡{self.jugador_actual} ha llegado a 3 puntos y gana el juego!")
                return False

            self.alternar_turno()
            return True


    def iniciar_juego(self):
        print("¡BIENVENIDAS AL JUEGO DE ADIVINANZA DE PLAYAS ESPAÑOLAS!")
        print("Escribe 'pista' para ayuda o 'salir' para terminar el juego.\n")


        random.shuffle(self.playas) # que salga una playa aleatoria
        print(f"Comienza el turno de {self.jugador_actual}")

        for playa in self.playas: # Recorremos las playas de la lista

            resultado = self.jugar_ronda(playa) #jugamos una ronda

            if resultado == False:  #si el resultado es False, significa que el jugador quiere sali
                break  
    

        print("\n--- MARCADOR FINAL---")
        for jugador, punto in self.puntos.items():
            print(f"{jugador}: {punto}")

        if self.puntos["Jugador 1"] > self.puntos["Jugador 2"]:
            print("GANADOR: Jugador 1")

        else:
            print("GANADOR: Jugador 2")
       
```

```python

juego = AdivinarPlayas() #instancia de clase
juego.iniciar_juego()  #llamamos al método
```
