# 🎉 Trivial: "Dímelo a mí, guapi!" 🎉

Este juego de preguntas permite adivinar frases famosas de la televisión. 

Puedes responder con **A, B o C**, o escribir `"salir"` para terminar el juego en cualquier momento. Tus respuestas correctas se iran sumando hasta el final.


```python
## JUEGO DEL TRIVIAL

# DICCIONARIOS
    # Creamos nuestras preguntas en formato diccionario, para poder ir accediendo a sus claves-valor a medida que avancemos el juego.
trivial = [{'pregunta' : 'PREGUNTA 1. \n ¿Quién dijo la frase: "¡Yo por mi hija, MA-TO!"?', 
              'opciones': 'OPCIONES: \n A) María Patiño \n B) Belén Esteban \n C) Carmen Lomana', 
              'respuesta' : 'b',
              'texto' : 'B) Belén Esteban'},
            {'pregunta' : 'PREGUNTA 2. \n ¿Quién dijo: "Y de mi vida hago lo que me sale del mismísimo coño, que lo sepáis todos"?',
             'opciones' : 'OPCIONES: \n A) La Veneno \n B) Carmen de Mairena \n C) María Jimenez', 
             'respuesta' : 'c',
             'texto' : 'C) María Jimenez'},
            {'pregunta' :  'PREGUNTA 3. \n ¿Quién dijo: "Si cada español pusiera una peseta..." cuando perdió un pendiente durante su actuación?',
             'opciones' : 'OPCIONES: \n A) Lola Flores \n B) Leticia Sabater \n C) Isabel Pantoja',
             'respuesta' : 'a',
             'texto' : 'A) Lola Flores'},
            {'pregunta' : 'PREGUNTA 4. \n ¿Quién dijo: “Tú y yo, Chelo… y te quiero” con tono misterioso?',
             'opciones' : 'OPCIONES: \n A) María Patiño \n B) Bárbara Rey \n C) Marta López',
             'respuesta' : 'b',
             'texto' : '\n B) Bárbara Rey'},
            {'pregunta' : 'PREGUNTA 5. \n ¿Quién dijo con gran indignación: “¿Y qué currículum tiene esta tarántula”?',
             'opciones' : 'OPCIONES: \n A) Belén Esteban \n B) Ylenia \n C) La Veneno',
             'respuesta' : 'c',
             'texto' : '\n C) La Veneno'},
            {'pregunta' : 'PREGUNTA 6. \n ¿Quién dijo: “Idiomas, querida” para presumir de su nivel de inglés?',
             'opciones' : 'OPCIONES: \n A) Aramís Fuster \n B) Tamara Falcó \n C) Isabel Preysler',
             'respuesta' : 'a',
             'texto' : '\n A) Aramís Fuster'}]

# MENSAJE DE BIENVENIDA
print('==============================')
print('BIENVENIDA A 🎉DÍMELO A MÍ, GUAPI!🎉 : \n El trivial más flamenco de adivinanzas! \n Te lanzaremos frases famosas leyendas del folklore televisivo. \n Tu misión: adivinar quién lo dijo o completar la frase. \n Si aciertas, ¡puntos! 😎 \n Si fallas… tranqui, que no pasa ná, aprendes pa la próxima! 💃\n COMENZAMOS!')

# CONTADOR
puntuacion = 0 # creamos un contador de puntos

# COMPROBACIONES
for pregunta in trivial: # vamos iterando las preguntas del trivial
        print('\n______________________________\n')
        print(pregunta["pregunta"]) # pedimos que imprima el texto de la pregunta accediendo al valor de esa clave
        print('\n_______')
        print(pregunta["opciones"]) # pedimos que imprima las opciones de respuesta accediendo al valor de esa clave

        while True: # bucle para repetir hasta que la respuesta sea válida
            respuesta = input("Indica tu respuesta usando 'A', 'B' o 'C' (o escribe 'salir' para terminar): ").lower().strip() # solicitamos la respuesta del jugador
            if respuesta == "salir": # para salir el bucle si el usuario decide terminar
                print()
                print("👋 Has decidido terminar el juego. ¡Hasta la próxima!")
                break
            elif respuesta not in ['a', 'b', 'c']: # para asegurarnos de que escribe la letra correcta, si su respuesta no es 'a', 'b' o 'c'
                print()
                print("⚠️ Ay chiquilla, que te has equivocao! Haz el favor de responder con 'A', 'B' o 'C'")
                continue # vuelve a pedir la respuesta
            break # si la respuesta es válida, salimos del bucle while

        if respuesta == "salir":  # rompe también el bucle for para terminar el juego
            break

        if respuesta == pregunta['respuesta']: # para validar las respuestas correctas
            print()
            print(f"✅ Efectivamente! La respuesta era {pregunta['texto']}")
            puntuacion += 1 # para sumar puntos
        else: # en caso de que no acierte, pasamos al else que solo se imprime, no suma puntos
            print()
            print(f"❌ Ohhh, no! Te has equivocado. La respuesta era {pregunta['texto']}")

# PUNTUACIÓN FINAL
print('\n==============================')
print(f"Tu puntuación final es: {puntuacion}/{len(trivial)}") # divido el número de respuestas correctas por el numero total de preguntas

# MENSAJES DE PUNTUACIÓN
if puntuacion < 2:
    print("Sigue estudiando, querida 😅")
elif puntuacion < 4:
    print("Not bad! 👍")
else:
    print("Eres una máquina de la cultura popular! 🎉")
print('==============================')
```
