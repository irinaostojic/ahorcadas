# ‚ú® Trivial: "D√≠melo a m√≠, guapi!" ‚ú®

Este juego de preguntas permite adivinar frases famosas de la televisi√≥n. 

Puedes responder con **A, B o C**, o escribir `"salir"` para terminar el juego en cualquier momento. Tus respuestas correctas se iran sumando hasta el final.


```python
## JUEGO DEL TRIVIAL

# DICCIONARIOS
    # Creamos nuestras preguntas en formato diccionario, para poder ir accediendo a sus claves-valor a medida que avancemos el juego.
trivial = [{'pregunta' : 'PREGUNTA 1. \n ¬øQui√©n dijo la frase: "¬°Yo por mi hija, MA-TO!"?', 
              'opciones': 'OPCIONES: \n A) Mar√≠a Pati√±o \n B) Bel√©n Esteban \n C) Carmen Lomana', 
              'respuesta' : 'b',
              'texto' : 'B) Bel√©n Esteban'},
            {'pregunta' : 'PREGUNTA 2. \n ¬øQui√©n dijo: "Y de mi vida hago lo que me sale del mism√≠simo co√±o, que lo sep√°is todos"?',
             'opciones' : 'OPCIONES: \n A) La Veneno \n B) Carmen de Mairena \n C) Mar√≠a Jimenez', 
             'respuesta' : 'c',
             'texto' : 'C) Mar√≠a Jimenez'},
            {'pregunta' :  'PREGUNTA 3. \n ¬øQui√©n dijo: "Si cada espa√±ol pusiera una peseta..." cuando perdi√≥ un pendiente durante su actuaci√≥n?',
             'opciones' : 'OPCIONES: \n A) Lola Flores \n B) Leticia Sabater \n C) Isabel Pantoja',
             'respuesta' : 'a',
             'texto' : 'A) Lola Flores'},
            {'pregunta' : 'PREGUNTA 4. \n ¬øQui√©n dijo: ‚ÄúT√∫ y yo, Chelo‚Ä¶ y te quiero‚Äù con tono misterioso?',
             'opciones' : 'OPCIONES: \n A) Mar√≠a Pati√±o \n B) B√°rbara Rey \n C) Marta L√≥pez',
             'respuesta' : 'b',
             'texto' : '\n B) B√°rbara Rey'},
            {'pregunta' : 'PREGUNTA 5. \n ¬øQui√©n dijo con gran indignaci√≥n: ‚Äú¬øY qu√© curr√≠culum tiene esta tar√°ntula‚Äù?',
             'opciones' : 'OPCIONES: \n A) Bel√©n Esteban \n B) Ylenia \n C) La Veneno',
             'respuesta' : 'c',
             'texto' : '\n C) La Veneno'},
            {'pregunta' : 'PREGUNTA 6. \n ¬øQui√©n dijo: ‚ÄúIdiomas, querida‚Äù para presumir de su nivel de ingl√©s?',
             'opciones' : 'OPCIONES: \n A) Aram√≠s Fuster \n B) Tamara Falc√≥ \n C) Isabel Preysler',
             'respuesta' : 'a',
             'texto' : '\n A) Aram√≠s Fuster'}]

# MENSAJE DE BIENVENIDA
print('==============================')
print('BIENVENIDA A ‚ú®D√çMELO A M√ç, GUAPI!‚ú®: \n El trivial m√°s flamenco en el que te lanzaremos frases famosas leyendas del folklore televisivo. \n Tu misi√≥n: adivinar qui√©n lo dijo. \n Si aciertas, puntos! üòé \n Si fallas‚Ä¶ tranqui, que no pasa n√°, aprendes pa la pr√≥xima! üíÉüèº\n COMENZAMOS!')

# CONTADOR
puntuacion = 0 # creamos un contador de puntos

# COMPROBACIONES
for pregunta in trivial: # vamos iterando las preguntas del trivial
        print('\n______________________________\n')
        print(pregunta["pregunta"]) # pedimos que imprima el texto de la pregunta accediendo al valor de esa clave
        print('\n_______')
        print(pregunta["opciones"]) # pedimos que imprima las opciones de respuesta accediendo al valor de esa clave

        while True: # bucle para repetir hasta que la respuesta sea v√°lida
            respuesta = input("Indica tu respuesta usando 'A', 'B' o 'C' (o escribe 'salir' para terminar): ").lower().strip() # solicitamos la respuesta del jugador
            if respuesta == "salir": # lo utilizamos para salir el bucle si el usuario decide terminar
                print()
                print("üëãüèº Has decidido terminar el juego. Nos vemos cuando te sientas preparada!")
                break
            elif respuesta not in ['a', 'b', 'c']: # para asegurarnos de que escribe la letra correcta, si su respuesta no es 'a', 'b' o 'c'
                print()
                print("‚ö†Ô∏è Ay chiquilla, que te has equivocao! Haz el favor de responder con 'A', 'B' o 'C'")
                continue # vuelve a pedir la respuesta
            break # si la respuesta es v√°lida, salimos del bucle while

        if respuesta == "salir":  # rompe tambi√©n el bucle for para terminar el juego
            break

        if respuesta == pregunta['respuesta']: # para validar las respuestas correctas
            print()
            print(f"‚úÖ Efectivamente! La respuesta era {pregunta['texto']}") # pregunta es todo el diccionario de esa pregunta, as√≠ que printeo accediento al valor 'texto' dentro de mi diccionario
            puntuacion += 1 # para sumar puntos
        else: # en caso de que no acierte, pasamos al else que solo se imprime, no suma puntos
            print()
            print(f"‚ùå Ohhh, no! Te has equivocado. La respuesta correcta era {pregunta['texto']}")

# PUNTUACI√ìN FINAL
print('\n==============================')
print(f"TU PUNTUACI√ìN FINAL ES: {puntuacion}/{len(trivial)}") # divido el n√∫mero de respuestas correctas por el numero total de preguntas

# MENSAJES DE PUNTUACI√ìN
if puntuacion < 2:
    print("Sigue estudiando, querida üòÖ")
elif puntuacion < 4:
    print("Not bad! üëçüèº")
else:
    print("Eres una m√°quina de la cultura popular! üéâ")
print('==============================')
```
