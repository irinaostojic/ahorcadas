```python
import random  # libreria para combinaciones aleatorias

# Palabras de terror escogidas
palabras_terror = [ #Se usa lista porque es fácil de recorrer, se puede modificar y funciona con random.choice().
    "espectro", "cripta", "ataud", "tinieblas", "cementerio", "maldicion",
    "fantasma", "exorcismo", "calavera", "posesion", "hechizo", "bruja",
    "ultratumba", "demonio", "sepulcro"
]
#queria que las frases fueran cambiando para hacer mas divertida la experiencia.
# Diccionario de frases de terror.#Con un dict puedes acceder mas rapido a ellas segun la situacion del juego.
frases_terror = {
    "bienvenida": [
        "Bienvenida... al ultimo juego de tu vida.",
        "La noche es oscura. Bienvenida, esta puede ser la ultima palabra que adivines.",
        "Has entrado. ¡Bienvenida! No podras salir hasta que ganes."
    ],
    "acierto": [
        "Si... pero ellos ya saben que estas aqui.",
        "Acertaste... pero algo se mueve en las sombras.",
        "Muy bien... por ahora sobrevives."
    ],
    "fallo": [
        "Has fallado... y la soga aprieta un poco mas.",
        "No... y algo se arrastra hacia ti.",
        "Ese error te acerca al final."
    ],
    "victoria": [
        "Has sobrevivido... pero no por mucho tiempo.",
        "Ganaste... o eso crees.",
        "La palabra era tu salvacion... esta vez."
    ],
    "derrota": [
        "Has perdido... y ahora eres uno de ellos.",
        "No lograste escapar... la oscuridad te envuelve.",
        "Caiste... y nadie vendra a ayudarte."
    ]
}

# Seleccion aleatoria de palabra
palabra = random.choice(palabras_terror)

# Mensajes iniciales
print(random.choice(frases_terror["bienvenida"]))#elige frase aleatoria de la categoria bienvenida.
print(f"Tu palabra contiene {len(palabra)} letras: {'_ ' * len(palabra)}")#se imprimen guines bajos por cada letra de la longitud de tu palabra.
print(palabra) #la imprimimos para ver si nos funciona el código

# Etapas del juego del ahorcado
ahorcado = [
    """
     ------
     |    |
     |
     |
     |
    _|_
    """,
    """
     ------
     |    |
     |    O
     |
     |
    _|_
    """,
    """
     ------
     |    |
     |    O
     |    |
     |
    _|_
    """,
    """
     ------
     |    |
     |    O
     |   /|
     |
    _|_
    """,
    """
     ------
     |    |
     |    O
     |   /|\\
     |
    _|_
    """,
    """
     ------
     |    |
     |    O
     |   /|\\
     |   /
    _|_
    """,
    """
     ------
     |    |
     |    O
     |   /|\\
     |   / \\
    _|_
    """
]

# Variables de control del juego
letras_correctas = set()#Son practicos aqui porque no permiten duplicados.
letras_incorrectas = set()#por ello para comprobar si una letra fue usada es mas rapido que una lista.
letras_usadas = set()
intentos = len(ahorcado) - 1 #Cuantos intentos tiene nuestra lista de ahorcado? 7 (tienes 6 intentos reales) y restas 1 porque el primero es el dibujo inicial.
fallos = 0 #contador de errores,cada vez que fallas suma 1.

# Configuracion de jugadores y marcador de puntos para cada jugador
jugadores = ["JUGADOR 1", "JUGADOR 2"]
puntos = {jugadores[0]: 0, jugadores[1]: 0}  # marcador inicial 
turno = 0  # empieza el Jugador 1
print(f"Tienes {intentos} intentos para acertar la palabra. Vivir o morir. Tu decides.")

# Bucle principal del Juego del Ahorcado
while True:#porque el juego no sabe cuantos turnos habrá y solo termina hasta que se cumple una condicion Break por victoria o derrota.

    # Nos muestra el dibujo inicial del ahorcado y nos muestra el turno del jugador
    jugador_actual = jugadores[turno]#turno es una variable que guarda 0 o 1 dependiendo del jugador.
    print(f"\nTURNO DE {jugador_actual}")#es una mamnera estetica de escrbir de quien es el turno.
    print(ahorcado[fallos])#muestra el dibujo del ahorcado segun la cantidad de fallos.

    # Mostrar estado actual de la palabra
    for letra in palabra:
        if letra.lower() in letras_correctas:
            print(letra, end=" ")# es un plantilla que siempre pone para pedirle a python que lo imprima todo en la misma linea.Si no hubiese un salto de linea por defecto.
        else:
            print("_", end=" ")#le dice a python: imprime un guion bajo seguido por un espacio en la misma linea para las letras que no han adivinado.
    print()

    print(f"Letras usadas: {', '.join(sorted(letras_usadas)) if letras_usadas else 'Ninguna'}")#.sorted ordenada alfabeticamente las letras que ya se probaron.
#.join(..) las une en un solo texto separadas por comas.
    # Introduccion de letra
    letra = input("Introduce una letra: ").strip().lower() #pide una letra al jugador y quita los espacios y la convierte a minuscula para que no haya errores.

    # Validaciones a las entradas de letra
    if len(letra) != 1 or not letra.isalpha():#le dice a Python: si la medida del caracter no es la de una sola letra 
        print(f"{letra} no es un caracter valido. Prueba de nuevo")#o si lo que se escribio no es una letra(rechaza todo lo que no sea una sola letra de alfabeto)
        continue # se salta los pasos posteriores y vuelve a empezar el bucle

    if letra in letras_usadas:#mira en letras usadas para comprobar.
        print(f"Ya has usado la letra '{letra}'. Prueba con otra.")
        continue# si se cumple ignora pasos posteriores y vuelve a empezar.

    letras_usadas.add(letra)# añade la letra a letras usadas.

    # Comprobacion de letra
    if letra in palabra.lower():
        letras_correctas.add(letra)
        print(f"{random.choice(frases_terror['acierto'])}. La '{letra}' esta en la palabra buscada")#elige al azar una frase de la seccion aciertos e imprime la letra que el jugador adivinó.
    else:
        letras_incorrectas.add(letra)#añade letra a letras incorrectas.
        fallos += 1# y suma un fallo
        print(f"{random.choice(frases_terror['fallo'])}. La '{letra}' no esta en la palabra buscada")#elige al azar una frase de la seccion freases de terror e imprime la letra que el jugador adivinó.
        print(f"Te quedan {intentos - fallos} intentos")# y aqui resta fallos de los intentos que queden
        turno = 1 - turno  # cambiar turno solo si falla
        if fallos < intentos:# si no has agotado tus intentos imprime y continua.
            print(f"Ahora es el turno de {jugadores[turno]}")

    # Condicion de aciertos en el Juego
    if letras_correctas >= set(palabra.lower()):#si la variable correctas contiene todas las letras de la palabra.
        print(random.choice(frases_terror["victoria"]))#elige al azar una frase de la seccion victoria
        print(f"¡{jugadores[turno]} ERES EL GANADOR DEL JUEGO DEL AHORCADO! La palabra era: {palabra}")#mira el turno de que jugador es y le da la victoria
        puntos[jugadores[turno]] += 1  # va sumando puntos al que acierta
        print(f"Marcador actual: {jugadores[0]}: {puntos[jugadores[0]]} | {jugadores[1]}: {puntos[jugadores[1]]}")#imprime el marcador con los nombres y puntos de cada jugador.
        break

    # Condicion de fallos en el Juego
    if fallos == intentos:#los fallos son los mismos que los intentos.
        print(f"{random.choice(frases_terror['derrota'])}. La palabra era: {palabra}")#elige al azar una frase de la seccion frases de terror e imprime la letra que el jugador adivinó.
        ganador = jugadores[1 - turno]#asi se alternan los turnos entre los dos jugadores.
        print(f"¡{ganador} ES EL GANADOR DEL JUEGO DEL AHORCADO!")#la variable ganador guarda el truno del ganador para luego usarla para imprimir y sumar puntos
        puntos[ganador] += 1  # sumar punto al ganador
        print(f"El Marcador Final es: {jugadores[0]}: {puntos[jugadores[0]]} | {jugadores[1]}: {puntos[jugadores[1]]}")#imprime el marcador con los nombres y puntos de cada jugador.
        break
```
