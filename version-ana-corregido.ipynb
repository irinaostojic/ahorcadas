```python
import random

palabras = ['lógica', 'proactividad', 'creatividad', 'resiliencia',
            'comunicación', 'python', 'programación', 'datos']        #tenemos que meter más palabras

palabra = random.choice(palabras)

print("¡¡¡BIENVENIDA AL JUEGO DEL AHORCADO!!!")
print(palabra)                                                      # esto lo tenemos que quitar, es solo para ver que palabra es e ir probando
print(f"Tu palabra contiene {len(palabra)} letras: {'_ ' * len(palabra)}")    #poner más print, hacerlos más complejos

ahorcado = [
    """
     ------
     |    |
     |
     |
     |
    _|_
    """,
    """
     ------
     |    |
     |    O
     |
     |
    _|_
    """,
    """
     ------
     |    |
     |    O
     |    |
     |
    _|_
    """,
    """
     ------
     |    |
     |    O
     |   /|
     |
    _|_
    """,
    """
     ------
     |    |
     |    O
     |   /|\\
     |
    _|_
    """,
    """
     ------
     |    |
     |    O
     |   /|\\
     |   /
    _|_
    """,
    """
     ------
     |    |
     |    O
     |   /|\\
     |   / \\
    _|_
    """
]

# Conjuntos (si lo hacemos con conjuntos por la propia definición del mismo, es más fácil que tratar que con listas)
letras_correctas = set()
letras_incorrectas = set()
letras_usadas = set()

# Diccionario de vocales acentuadas y sin acentuar
acentos = {
    'á': 'a', 'é': 'e', 'í': 'i', 'ó': 'o', 'ú': 'u',
    'à': 'a', 'è': 'e', 'ì': 'i', 'ò': 'o', 'ù': 'u'
}

# Intentos
intentos = len(ahorcado) - 1
fallos = 0
print(f"Tienes {intentos} intentos para acertar la palabra. ¡Ánimo!")
```

```python


while True:
    print(ahorcado[fallos])

    # Bucle inicial de iteración. Aquí nos muestra la letra o guión. También aquí le decimos a python que nos reconozca la vocal aunque nos equivoquemos en el acento
    for letra_original in palabra:
        letra_sin_acento = acentos.get(letra_original.lower(), letra_original.lower())
        if letra_sin_acento in letras_correctas:
            print(letra_original, end=" ")
        else:
            print("_", end=" ")
    print()

    letra = input("Introduce una letra: ").strip().lower()
    letra = acentos.get(letra, letra)  # normalizamos la entrada

    if len(letra) != 1 or not letra.isalpha():
        print(f"{letra} no es un carácter válido. Prueba de nuevo")
        continue

    if letra in letras_usadas:
        print(f"Ya has usado la letra '{letra}'. Prueba con otra.")
        continue

    letras_usadas.add(letra)

    # Comprobamos letra quitando acentos a la palabra
    if letra in [acentos.get(c, c) for c in palabra.lower()]:
        letras_correctas.add(letra)
        print(f"¡Bien! La letra '{letra}' está en la palabra.")
    else:
        letras_incorrectas.add(letra)
        fallos += 1
        print(f"La letra '{letra}' no está en la palabra")
        print(f"Te quedan {(len(ahorcado) - 1 - fallos)} intentos")


    # Ganamos en el juego
    if letras_correctas >= set(acentos.get(c, c) for c in palabra.lower()):
        print(f"¡ENHORABUENA! HAS ACERTADO LA PALABRA: {palabra}")
        break

    # Perdemos en el juego. Hemos consumido todas las vidas
    if fallos == intentos:
        print(f"¡GAME OVER! La palabra era: {palabra}")
        print(ahorcado[fallos])
        break
```
